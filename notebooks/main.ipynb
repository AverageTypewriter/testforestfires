```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('forestFireDataset.csv', header = 1)
```

```python
dataset.head()
```

```python
dataset.info()
```

### Data Cleaning

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:121, "Region"]=0
dataset.loc[122:, "Region"]=1
```

```python
dataset.info()
df = dataset
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
## Fix spaces in column names
df.columns = df.columns.str.strip()
df.columns
```

## Change the required columns as integer data type

```python
df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']]=df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.columns
```

```python
### Changing the other columns to float datatye

objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.to_csv('forestFireDataset.csv', index=False)
```

### Exploratory Data Analysis

```python
df_copy=df.drop(['day', 'month', 'year'], axis=1)
```

```python
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'), 0, 1)
```

```python
# Plot the density plot for all features

df_copy.hist(bins = 50, figsize= (20,15))
plt.show()
```

```python
## Percentage for Pie Chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
classlabes = ["Fire", "Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage, labels = classlabes, autopct='%1.1f%%')
plt.title("Pie chart of classes")
plt.show()
```
